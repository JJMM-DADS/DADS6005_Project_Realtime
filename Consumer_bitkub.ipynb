```python
# Consumer
from kafka import KafkaConsumer
import json
import pandas as pd
consumer = KafkaConsumer(
 'jjmm_bitkub', # replace with your topic
 bootstrap_servers=['"xxxx":9092']) #replace "xxxx" with kafka server
i=0
for message in consumer:
    message = message.value.decode("utf-8")
    string = message.replace("'",'"')
    jsonData = json.loads(string)
    df=pd.DataFrame([jsonData])
    if i==0:
        df.to_csv("record_bitkub.csv",index=False)
        print(df[["timestamp","last"]])
    else:
        df1=pd.read_csv("record_bitkub.csv")
        all_df=pd.concat([df1,df])
        all_df.to_csv("record_bitkub.csv",index=False)
        print(all_df[["timestamp","last"]])
    print("--------***-----\n")
    i=i+1
```

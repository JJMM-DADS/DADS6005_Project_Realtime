```python
# Consumer
from kafka import KafkaConsumer
import json 
import pandas as pd
consumer = KafkaConsumer(
 'jjmm-stream',
 bootstrap_servers=['ec2-13-229-46-113.ap-southeast-1.compute.amazonaws.com:9092'])
i=0
for message in consumer:
    message = message.value.decode("utf-8")
    string = message.replace("'",'"')
    jsonData = json.loads(string)
    df=pd.DataFrame([jsonData])
    if i==0:
        df.to_csv("record.csv",index=False)
        print(df[["timestamp","lastPrice","rate"]])
    else:
        df1=pd.read_csv("record.csv")
        all_df=pd.concat([df1,df])
        all_df.to_csv("record.csv",index=False)
        print(all_df[["timestamp","lastPrice","rate"]])
    print("--------***-----\n")
    i=i+1
```
